```python
import pandas as pd
```

```python
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRSbxW53joz0J2_sB9wz_k-QgZT9aPSvET2DXjb8ILMMSJVEAMyaTE-BKRZW96GRHNmWrn_JyAn-n0q/pub?gid=1509851203&single=true&output=csv")
```

```python
top_10_artists = df.groupby('artist')['streams'].sum().nlargest(10).index
final_df = pd.DataFrame(columns=['Artista', 'Géneros'])

for artist_name in top_10_artists:
    results = sp.search(q=f'artist:{artist_name}', type='artist')
    artist = results['artists']['items'][0]
    genres = artist['genres']

    temp_df = pd.DataFrame({'Artista': [artist_name], 'Géneros': [genres]})
    final_df = pd.concat([final_df, temp_df], ignore_index=True)

final_df
```
